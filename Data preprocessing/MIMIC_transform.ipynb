```python
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/1129medWithATC2/p1.csv")
print(df.shape)
print(df.head())
```

```python
df = df.dropna()
print(df.head(20))

```

```python
df["date"] = df.apply(
    lambda row: pd.date_range(row["STARTDATE"], row["ENDDATE"]),
    axis=1
)
df = (
    df.explode("date", ignore_index=True)
    .drop(columns=["STARTDATE", "ENDDATE"])
)

print(df.head(20))
```

```python
one_hot = pd.get_dummies(df['string_field_0'])
```

```python
df = df.join(one_hot)
print(df.head(20))
```

```python
df.to_csv("/content/drive/My Drive/1129med_single_day_onehot.csv", sep='\t')
```
